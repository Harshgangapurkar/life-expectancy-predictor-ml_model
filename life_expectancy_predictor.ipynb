# Training

```python
y = df[['Lifeexpectancy']]
x = df.drop(columns = ['Lifeexpectancy'])
```

```python
x = np.array(x).astype('float32')
y = np.array(y).astype('float32')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
!pip install xgboost
```

```python
import xgboost as xgb
model = xgb.XGBRegressor(objective = 'reg:squarederror',learing_rate = 0.1,max_depth = 30,n_esimators = 100)
model.fit(x_train,y_train)
```

```python
import joblib
joblib.dump(model, "life_expectancy_model.pkl")
```

# Evaluation

```python
accuracy = model.score(x_test,y_test)
print(accuracy)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(x_test)

# Plot actual vs predicted
plt.figure(figsize=(10,6))
# Convert y_test to 1-dimensional array
sns.scatterplot(x=y_test.flatten(), y=y_pred)
plt.xlabel('Actual Life Expectancy')
plt.ylabel('Predicted Life Expectancy')
plt.title('Actual vs Predicted Life Expectancy')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
```

```python
import gradio as gr
import pandas as pd
import joblib

# Load your model
model = joblib.load("life_expectancy_model.pkl")  # path to your saved model

# Prediction function
def predict_life_expectancy(
    Year, AdultMortality, infantdeaths, Alcohol, percentageexpenditure,
    HepatitisB, Measles, BMI, under_fivedeaths, Polio, Totalexpenditure,
    Diphtheria, HIV_AIDS, GDP, Population, thinness_19, thinness_5,
    Incomecomposition, Schooling, Status
):
    data = {
        'Year': [Year],
        'AdultMortality': [AdultMortality],
        'infantdeaths': [infantdeaths],
        'Alcohol': [Alcohol],
        'percentageexpenditure': [percentageexpenditure],
        'HepatitisB': [HepatitisB],
        'Measles': [Measles],
        'BMI': [BMI],
        'under-fivedeaths': [under_fivedeaths],
        'Polio': [Polio],
        'Totalexpenditure': [Totalexpenditure],
        'Diphtheria': [Diphtheria],
        'HIV/AIDS': [HIV_AIDS],
        'GDP': [GDP],
        'Population': [Population],
        'thinness1-19years': [thinness_19],
        'thinness5-9years': [thinness_5],
        'Incomecompositionofresources': [Incomecomposition],
        'Schooling': [Schooling],
        'Status_Developed': [Status == "Developed"],
        'Status_Developing': [Status == "Developing"]
    }
    df = pd.DataFrame(data)
    prediction = model.predict(df)
    return f"Predicted Life Expectancy: {prediction[0]:.2f} years"

# Interface
interface = gr.Interface(
    fn=predict_life_expectancy,
    inputs=[
        gr.Number(label="Year"),
        gr.Number(label="Adult Mortality"),
        gr.Number(label="Infant Deaths"),
        gr.Number(label="Alcohol"),
        gr.Number(label="Percentage Expenditure"),
        gr.Number(label="Hepatitis B"),
        gr.Number(label="Measles"),
        gr.Number(label="BMI"),
        gr.Number(label="Under-Five Deaths"),
        gr.Number(label="Polio"),
        gr.Number(label="Total Expenditure"),
        gr.Number(label="Diphtheria"),
        gr.Number(label="HIV/AIDS"),
        gr.Number(label="GDP"),
        gr.Number(label="Population"),
        gr.Number(label="Thinness 1-19 Years"),
        gr.Number(label="Thinness 5-9 Years"),
        gr.Number(label="Income Composition of Resources"),
        gr.Number(label="Schooling"),
        gr.Radio(["Developed", "Developing"], label="Status")
    ],
    outputs="text",
    title="Life Expectancy Predictor"
)

interface.launch()
```
